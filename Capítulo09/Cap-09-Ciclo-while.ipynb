**Ciclo while**

```python
i = 1
while(i <= 5):
  print("Bienvenido a Python")
  i = i+1

print("Fin programa")
```

**Ejemplo de ciclo while con visualización**

```python
i = 1
while(i <= 5):
  print(i)
  i = i+1

print("Fin programa")
```

**Ciclos controlados por valor centinela**

```python
nombre = ""
while(nombre != "Daniel"):
  nombre = input("Ingrese su nombre: ")

print("Bienvenido")
```

**Break**

```python
while(True):
  capital = input("Ingrese la capital de Perú: ")
  if(capital == "Lima"):
    print("Felicidades ganaste la trivia")
    Break

print("Fin programa")
```
